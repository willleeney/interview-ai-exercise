```python
from ai_exercise.constants import SETTINGS, chroma_client
from ai_exercise.retrieval.vector_store import create_collection
from ai_exercise.llm.embeddings import openai_ef
from ai_exercise.loading.document_loader import get_json_data, build_docs, split_docs, add_documents

# Create the collection
collection = create_collection(chroma_client, openai_ef, SETTINGS.collection_name)
print(f"Initial collection count: {collection.count()}")

# Get and process the documents
json_data = get_json_data()
print(f"JSON data loaded, entries: {len(json_data)}")

documents = build_docs(json_data)
print(f"Built documents: {len(documents)}")

split_documents = split_docs(documents)
print(f"Split documents: {len(split_documents)}")

# Add documents to the collection
add_documents(collection, split_documents)
print(f"Final collection count: {collection.count()}")
```
